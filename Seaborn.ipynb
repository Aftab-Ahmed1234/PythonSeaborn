# Barplot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
students=["Aftab","Arslan","Abdullah"]
marks=[95,85,90]

sns.barplot(x=students,y=marks)
plt.title("Student Marks",color="blue", size=20)
plt.xlabel("Students", color="red", size=15)
plt.ylabel("Marks", color="red", size=15)
plt.show()
```

# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
x=[90,54,78,96,84,100]
sns.histplot(x)
plt.title("Marks Distribution")
plt.show()
```
